# Simple Cicada Model

While we get the real code straightened out, we can use this model to gain insight into how groups of computers emitting and listening to sounds actually might behave. This will be useful in determining spatial patterns for the physical demonstration, in working out the visualizations, and in creating an intro presentation.

**Caveats:**
* **Pitches are modeled as a single frequency.** This is not possible with the crappy laptop speakers even if we tried. Further, we're thinking of switching from tones to piano notes. We could extend this to model tones a gaussians or something.
* **Intensities are summed.** In reality, sound intensities do not sum linearly given that it is unlikely that the two sound waves are perfectly in phase. We could extend this model to include sounds as waves.
* **Pitch and loudness are quantities really based around human perception of sound.** It's not technically accurate to describe what the algorithm is doing, but it will be useful to talk about it this way to the kids.
* **Pitches are recorded by timepoint and not by agent.** In the future it may be useful to understand the history of a particular agent, but for now this should be sufficient for visualizing the dynamic.

**Fun Extensions:**
* Make each cicada's pitch jump be proportional to the size of the gap between the two pitches. That might be necessary for convergence. Looking at it's behaviour thus far, it's not converging.
* This is sort of like a (spatially) weighted iterative relaxation method. Could we apply this to solving something cool?

```python
from math import sqrt
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
# frequency range of typical laptop speaker
f_low = 150 # hz
f_high = 5000 # hz
f_range = f_high - f_low
population_size = 50
initial_intensity = 100 # this is something that could be individually determined (i.e., different speaker powers)
runtime = 2000
grid_size = (20,20)
# Step size for relaxation
df = 1
```

```python
class Cicada(object):
    _registry = []
    cic_count = 0
    
    def __init__(self, name, pitch, intensity, pos):
        self._registry.append(self)
        self.name = name
        self.pitch = pitch
        self.intensity = intensity
        self.sounds = np.zeros(f_range)
        self.pos = pos
        Cicada.cic_count += 1
   
    def displayCount(self):
        print "Total cicadas %d" % Cicada.cic_count

    def displayCicada(self):
        print "Name : ", self.name,  ", Pitch: ", self.pitch, ", intensity", intensity, ", pos", self.pos
```

```python
cicadas = [Cicada("cic" + str(i), np.random.randint(f_low, f_high), initial_intensity, 
                  (np.random.uniform(-grid_size[0],grid_size[0]), np.random.uniform(-grid_size[1], grid_size[1]))) for i in range(population_size)]
```

```python
plt.axis([-grid_size[0], grid_size[0], -grid_size[1], grid_size[1]])
plt.ion()

for cic in Cicada._registry:
    plt.scatter(cic.pos[0], cic.pos[1])
    plt.draw()
```

```python
def calc_intensity(my_pos, nei_pos, intensity):
    x1, y1 = my_pos
    x2, y2 = nei_pos
    r = sqrt((x2-x1)**2 + (y2-y1)**2)
    return intensity * (1/r**2)
```

```python
pitches = np.zeros((f_range,runtime))
for t in range(runtime):
    # iterate over all cicadas
    # compile a spectrum of "heard" sounds
    i = 0
    for cic in Cicada._registry:
        cic.sounds = np.zeros(f_range)
        for nei_cic in Cicada._registry:
            if cic.name != nei_cic.name:
                nei_intensity = calc_intensity(cic.pos, nei_cic.pos,nei_cic.intensity)
                cic.sounds[nei_cic.pitch-f_low] += nei_intensity
        # move closer in frequency to most intense pitch
        target_pitch = np.argmax(cic.sounds)
        pitch_gap = target_pitch - cic.pitch
        if pitch_gap > 0: cic.pitch += df
        elif pitch_gap < 0: cic.pitch -= df
        # record frequency of each cicada for plotting
        pitches[cic.pitch-f_low,t] += 1
        i += 1
```
